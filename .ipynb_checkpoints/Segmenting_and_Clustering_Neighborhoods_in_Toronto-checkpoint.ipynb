## Step 1: Transform data in the table on the Wikipedia page into the pandas dataframe

```python
#importing pandas library
import pandas as pd
```

<b>Retreiving Data Process

```python
#Read HTML file using pandas library and get the required table data
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
source = pd.read_html(url, header=0, attrs={'class': 'wikitable sortable'})
source[0]
```

```python
# assign the data output to a variable and rename column by removing the space from first column name
df = source[0]
df.columns=['PostalCode','Borough','Neighborhood']
df.head()
```

<b> Data Pre-Processing and Cleansing

```python
#Removing rows from dataframe having calue Not assigned in Borough column
df = df[df['Borough']!='Not assigned']
```

```python
#Checking row for having Not assigned value in Neighbour column if found then replace with Borough
if len(df[df['Neighborhood'] =='Not assigned'])>0:
    df.loc[df['Neighbourhood'] =='Not assigned' , 'Neighbourhood'] = df['Borough']
```

```python
#joining values of Neighborhood based on PostalCode and Borough
df = df.groupby(['PostalCode','Borough'], sort=False).agg( ', '.join)
```

```python
#Resetting index again
df = df.reset_index()
df.head()
```

```python
df.shape
```

## Step 2: Use the Geocoder package or the csv to retreive Latitude & Longitude.

<b>Getting the location data from given location using CSV file

```python
# Reading data from location
df_location = pd.read_csv('http://cocl.us/Geospatial_data')
df_location.head()
```

```python
#Renaming first column name by removing space
df_location.columns =['PostalCode','Latitude','Longitude']
df_location.head()
```

<b>Merging both dataframe into one

```python
#merge the original data with location information
df_data = df.merge(df_location, left_on='PostalCode', right_on='PostalCode')
df_data.head()
```

```python
df_data.shape
```

<b>Define Foursquare Credentials and Version

```python
CLIENT_ID = '' # your Foursquare ID
CLIENT_SECRET = '' # your Foursquare Secret
VERSION = '20180605' # Foursquare API version
LIMIT = 100 # limit of number of venues returned by Foursquare API
radius = 500 # define radius
```

<b>Function to get venues of all the neighborhoods in Toronto

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

<b> Call above function for all Toronto data.

```python
toronto_venues = getNearbyVenues(names=toronto_data['Neighborhood'],
                                   latitudes=toronto_data['Latitude'],
                                   longitudes=toronto_data['Longitude']
                                  )
```

<b>Size of the resulting dataframe

```python
print(toronto_venues.shape)
toronto_venues.head()
```

<b>Venues returned for each neighborhood.

```python
toronto_venues.groupby('Neighborhood').count()
```

<b>Analyze Each Neighborhood

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
```

```python
toronto_onehot.shape
```

<b>Group rows by neighborhood and by taking the mean of the frequency of occurrence of each category.

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped.head()
```

```python
toronto_grouped.shape
```

<b>Print each neighborhood along with the top 5 most common venues

```python
num_top_venues = 5

for hood in toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

<b>Function to sort the venues in descending order.

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

<b>Create the new dataframe and display the top 10 venues for each neighborhood.

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

### Cluster Neighborhoods

<b>Run *k*-means to cluster the neighborhood into 5 clusters.

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

<b>Dataframe that includes the cluster as well as the top 10 venues for each neighborhood.

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_data.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

toronto_merged.head() # check the last columns!
```

<b>Visualize the resulting clusters.

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
